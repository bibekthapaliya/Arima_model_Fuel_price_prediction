```python
tf=df['diesel']
tf.head(20)
```

```python
#Checking the stationery of time series


from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)
#Plot rolling statistics:
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
test_stationarity(tf)
```

```python
tf_log=np.log(tf)
tf_log.plot(figsize=(15,5))
```

```python
tf_log.head()
```

```python
test_stationarity(tf_log)
```

```python
tf_log_diff=tf_log-tf_log.shift(1)
plt.plot(tf_log_diff)
```

```python
tf_log_diff.dropna(inplace=True)
test_stationarity(tf_log_diff)
```

```python
from statsmodels.tsa.stattools import acf,pacf
```

```python
lag_acf = acf(tf_log_diff, nlags=20)
```

```python
lag_acf
```

```python
#Plot Acf
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(tf_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(tf_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
```

```python
lag_pacf = pacf(tf_log_diff, nlags=20, method='ols')
```

```python
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(tf_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(tf_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
#AR
model = ARIMA(tf_log, order=(1, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(tf_log_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-tf_log_diff)**2))
```

```python
#MA Model:
model = ARIMA(tf_log, order=(0, 1, 1))  
results_MA = model.fit(disp=-1)  
plt.plot(tf_log_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-tf_log_diff)**2))
```

```python
model = ARIMA(tf_log, order=(1, 1, 1))  
results_ARIMA = model.fit(disp=-1)  
plt.plot(tf_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-tf_log_diff)**2))
```
